# Examples of MPSE-tsne using mview.basic()

```python
import sys; sys.path.insert(0,"../")
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib notebook

import mview
```

We'll use our 1000 point 123 data set:

```python
path = 'data/123/input/'
X_true = np.genfromtxt(path+'spicy_rice_1000_123.csv', delimiter=',')
Y1 = np.genfromtxt(path+'spicy_rice_1000_1.csv', delimiter=',')
Y2 = np.genfromtxt(path+'spicy_rice_1000_2.csv', delimiter=',')
Y3 = np.genfromtxt(path+'spicy_rice_1000_3.csv', delimiter=',')
fig, ax = plt.subplots(1,3)
ax[0].plot(Y1[:,0],Y1[:,1],'o')
ax[1].plot(Y2[:,0],Y2[:,1],'o')
ax[2].plot(Y3[:,0],Y3[:,1],'o')
plt.show()
```

We run mpse-tsne using mview.basic() just as we do for mpse-mds. We just set 'visualization_method' = 'tsne'.

The line below will run mpse-tsne with standard parameters (the same as in mpse-mds, with perplexity = 30).

```python
mv = mview.basic([Y1,Y2,Y3], visualization_method = 'tsne', verbose=2)
```

```python
mv.plot_images()
```

```python
mv.plot_embedding()
```

To change the perplexity, we specify visualization_args = {'perplexity':###}."

```python
mv = mview.basic([Y1,Y2,Y3], visualization_method = 'tsne', visualization_args = {'perplexity':500}, verbose=2)
```

```python
mv.plot_images()
```

```python
mv.plot_embedding()
```

Other arguments work too, such as smart visualization (initialize using mds on the combined distances) or batch_size.

```python
mv = mview.basic([Y1,Y2,Y3], visualization_method='tsne', visualization_args={'perplexity':500}, 
                 smart_initialization=True, batch_size=50, verbose=2)
```

```python
mv.plot_images()
```

```python
mv.plot_embedding()
```
